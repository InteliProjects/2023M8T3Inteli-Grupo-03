```python
import time
import random

tempos_primeira_api = []
tempos_segunda_api = []

# Simulando a primeira API - Sistema Atual
def simular_primeira_api():
    await_time = random.uniform(0.2, 0.4)  # Usar uniform para gerar valores de ponto flutuante
    time.sleep(await_time)

# Simulando a segunda API - Sistema Novo
def simular_segunda_api():
    await_time = random.uniform(0.001, 0.1)  # Usar uniform para gerar valores de ponto flutuante
    time.sleep(await_time)

# Loop para realizar a simulação 100 vezes
for _ in range(10):
    # Realize uma simulação de chamada para a primeira API
    start_time = time.time()
    simular_primeira_api()
    end_time = time.time()
    response_time_api1 = end_time - start_time
    tempos_primeira_api.append(response_time_api1)

    # Realize uma simulação de chamada para a segunda API
    start_time = time.time()
    simular_segunda_api()
    end_time = time.time()
    response_time_api2 = end_time - start_time
    tempos_segunda_api.append(response_time_api2)

# Calcular médias e outras estatísticas
media_api1 = sum(tempos_primeira_api) / len(tempos_primeira_api)
media_api2 = sum(tempos_segunda_api) / len(tempos_segunda_api)
tempo_maximo_api1 = max(tempos_primeira_api)
tempo_maximo_api2 = max(tempos_segunda_api)

print("Estatísticas para o desempenho do Sistema Atual:")
print(f"Média de tempo de resposta: {media_api1:.2f} segundos")
print(f"Tempo máximo de resposta: {tempo_maximo_api1:.2f} segundos")

print("\nEstatísticas para o desempenho do Sistema Novo:")
print(f"Média de tempo de resposta: {media_api2:.2f} segundos")
print(f"Tempo máximo de resposta: {tempo_maximo_api2:.2f} segundos")
```

```python
import matplotlib.pyplot as plt

plt.scatter(tempos_primeira_api, tempos_segunda_api, label='Pontos de Dados', color='blue', marker='o')

# Adicionar rótulos e título
plt.xlabel('Tempo de Resposta do Sistema Atual (segundos)')
plt.ylabel('Tempo de Resposta do Sistema Novo (segundos)')
plt.title('Gráfico de Dispersão - Sistema Atual vs. Sistema Novo')

# Exibir o gráfico
plt.legend()
plt.show()
```
